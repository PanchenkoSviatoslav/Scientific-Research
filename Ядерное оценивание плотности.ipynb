```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.style.use('ggplot')
%matplotlib inline
from scipy import stats
from scipy.stats import norm
from math import sqrt
```

```python
def a(n):
    return 1/sqrt(n)
```

```python
def kernel(X_i, x, h):
    l = 2
    x_left = x - h*(l/2)
    x_right = x + h*(l/2)
    return 1/l if (x_left <= X_i and X_i <= x_right) else 0 
```

```python
def f_n(x, X):
    n = len(X)
    h = a(n)
    res = []
    for element in x:
        kernel_list = [kernel(X_i, element, h) for X_i in X]
        res.append(1/(n*h)*sum(kernel_list))
    return np.array(res)
```

```python
X = norm.rvs(size = 10000, random_state = 123)
```

```python
x = np.linspace(-5, 5, 101)
plt.plot(x, norm.pdf(x), '--')
plt.plot(x, f_n(x, X))
plt.show()
```
